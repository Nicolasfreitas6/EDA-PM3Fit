```python
#Carregando bibliotecas e datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

users = pd.read_csv("../data/users_dataset.csv")
user_plan = pd.read_csv("../data/user_plan.csv")
user_app = pd.read_csv("../data/user_app.csv")
activities = pd.read_csv("../data/activities_dataset.csv")
weekly_visits = pd.read_csv("../data/weekly_gym_visits_aggregated_dataset.csv")
challenges = pd.read_csv("../data/challenges_dataset.csv")
desafio_usuario = pd.read_csv("../data/desafio_usuario_dataset.csv")
```

```python
#Criando dicionário dos datasets
datasets = {
    'users': users,
    'user_plan': user_plan,
    'user_app': user_app,
    'activities': activities,
    'weekly_visits': weekly_visits,
    'challenges': challenges,
    'desafio_usuario': desafio_usuario
}
```

```python
#Verificando tipos de dados para confirmação
for name, df in datasets.items():
    print(f'\n==== {name} ====')
    print(df.dtypes)
```

```python
#Convertendo colunas de datas
user_plan['data_inicio'] = pd.to_datetime(user_plan['data_inicio'])

user_app['checkin'] = pd.to_datetime(user_app['checkin'])

activities['date'] = pd.to_datetime(activities['date'])

challenges['data_inicio'] = pd.to_datetime(challenges['data_inicio'])
challenges['data_fim'] = pd.to_datetime(challenges['data_fim'])

desafio_usuario['data_desafio'] = pd.to_datetime(desafio_usuario['data_desafio'])
```

```python
#Verificando novamente os tipos de dados após ajustes
for name, df in datasets.items():
    print(f'\n==== {name} ====')
    print(df.dtypes)
```

```python
#Verificando tamanho do dataset
df.shape
```

```python
#Verificando duplicatas dos dados novamente para confirmação   
for name, df in datasets.items():
    print(f'{name}: {df.duplicated().sum()} duplicadas')
```

```python
#Removendo duplicatas
user_plan = user_plan.drop_duplicates()

user_app = user_app.drop_duplicates()

desafio_usuario = desafio_usuario.drop_duplicates()
```

```python
#Verificando duplicatas novamente após correção
desafio_usuario.duplicated().sum()

user_app.duplicated().sum()

desafio_usuario.duplicated().sum()
```

Por não se tratarem de visitas a acadêmia com valores muito discrepantes a 7 (quantidade de dias da semana), não será excluído esses dados, pois irei assumir que certas pessoas podem fazer mais de 1 checkin por dia, por exemplo, aumentando esse número.
